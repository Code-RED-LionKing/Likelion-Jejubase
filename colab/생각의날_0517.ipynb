# 3. 그 데이터들을 가지고 어떻게 수익화(적자, 투자, 유료, 무료)를 실현하고 있나요?

뉴스기사등을 통해 이들의 상황이 어떤지에 대해서 파악해 보세요.(여론, 반응)

1. 거래소들의 기본 수익화 구조는 유저들이 거래를 하면서 지불하는 fee(수수료)
   국내 증권사의 수수료는 보통(0.25%)
   1000만원 매매시 25,000원


2. 증권사는 수수료 뿐만아니라 대출을 해주는 은행과 같다고 생각함.

   ex)신용융자 or 미수 -> 자신이 가진 금액(증거금)보다 더 많은 금액을 살 수 있음.
      예를들어 1000만원이 있다면 2500만원까지 주식을 살 수 있고 이에대한 이자를 
      납부 해야함. 
      
      단기간 주식을 정리를 할 경우 이자는 4~7% 대지만 90일이 도래하기 전
      주식을 처분 or 빌린 금액을 값아야함 61일 이상일 경우 이자 는 9%대
      
      주가가 일정가격 이하로 떨어질 경우 반대매매(청산 liquidation)발생. 매우위험
      
3. 수수료 무료(이벤트)를 통해서 타사증권을 사용하고있는 고객을 끌어당긴다.
